```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
nflstadium = pd.read_csv("/Users/colinmichael/Desktop/Data Science/DSC 550/nfl_stadiumdata.csv")
```

```python
#I pulled a data set from Kaggle.com on the National Football League (NFL) stadiums.
#There are 31 stadiums in the NFL, where weekly football games are played.
#The data set includes stadium name, capacity, location, surface type, roof type, home teams, and year opened.

#I want to see which variables are the most strongly related to stadium capacity.

nflstadium.sort_values(by=['Capacity'], ascending=False)
```

```python
#Graph 1.) Doing a scatterplot between stadium year built and capacity.
#Graph shows a slight negative relationship between year built and capacity.

theta = np.polyfit(nflstadium.Opened, nflstadium.Capacity, 1)
y_line = theta[1] + theta[0] * nflstadium.Opened
plt.scatter(nflstadium.Opened, nflstadium.Capacity, alpha=0.5)
plt.plot(nflstadium.Opened, y_line, 'r')
plt.show()
```
